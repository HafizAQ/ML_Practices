#Heirarchical Clustering in ML
Un-supervised Learning Approach
- Centroid based Clustering: K-Means Clustering, K-Mod Clustering  
- Distribution-based Clustering: Algorithms generate clusters from a particular probability distribution, e.g Guassian Metrix Model (GMM)
- Density-based Clustering: Group to gather data-points that are in high density concentrations and seperate low concentration regions e.g. Density-based Spatial Clustering of Application with Noise (DBSCAN)


####Hierarchical Clustering Types
- i) Agglomerative Clustering (Bottom-Up Approach)
- ii) Divisive Clustering (Top-Down Approach)

```python
#Hierarchical Divisive Clustering
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt

#Randomly chosen dataset
X = np.array([[1,2],[1,4],[1,0],[4,2],[4,4],[4,0]]) #Metrix

#Perform hierarchical clustering
Z = linkage(X, 'ward')

#Plot demdrogram
dendrogram(Z)

plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Point')
plt.ylabel('Distance')
plt.show()
```

#K-means Clustering - Introduction
- Unsupervised ML algorithm
- Groups the unlabled dataset into different clusters

```python
#Imp of K-Means Clustering in Python
#i) Import the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
#ii) Load the Dataset
X, y= load_iris(return_X_y=True)

print(X) #Metrics feature
print(y) #Target Vector
```

```python
#iii) Elbow Method (finding the ideal number of groups to divide the data into is a basic stage in any unsupervised algorithm)=> approach to find ideal K (no of clusters)
sse=[]
for k in range(1,11):
  km=KMeans(n_clusters=k, random_state=2)
  km.fit(X)
  sse.append(km.inertia_)

#Plot the Elbew graph to find the optimum number of cluster
sns.set_style("whitegrid")
g=sns.lineplot(x=range(1,11),y=sse)

g.set(xlabel="Number of Clusters", ylabel="Sum Squared  Error", title="Elbow Method")
plt.show()


#Elbow like sitation, k=2, k=3. So, we are considering K=3
```

```python
#Considering K=3
#iv) Build the Kmeans clustering model
kmeans = KMeans(n_clusters = 3, random_state = 2)
kmeans.fit(X) #giving feature metrix
```

```python
#v) Find the Cluster Center
print(kmeans.cluster_centers_)
#vi) Predict the Cluster Groups
pred = kmeans.predict(X)
print(pred)
```

```python
#Plot the cluster center with data points
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.scatter(X[:,0], X[:,1], c=pred, cmap=cm.Accent)
plt.grid(True)

for center in kmeans.cluster_centers_:
  center = center[:2]
  plt.scatter(center[0], center[1], c='black', marker='^')

plt.xlabel("Petal Length (cm) ")
plt.ylabel("Patel Width (cm) ")
plt.title("Cluster Centers 1")

plt.subplot(1,2,2)
plt.scatter(X[:,2], X[:,3], c=pred, cmap=cm.Accent)
plt.grid(True)

for center in kmeans.cluster_centers_:
  center = center[2:4]
  plt.scatter(center[0], center[1], c='black', marker='^')
plt.xlabel("Sepal Length (cm) ")
plt.ylabel("Sepal Width (cm) ")
plt.title("Cluster Centers 2")

plt.show()
```

#K-Mod Clustering
Gruop a set of data objects into a specified number of clusters, based on their **catagorical attributes**

Changes made for categorical data
- Replace Euclidean distance with Dissimilarity metrix (Computation in CS, evolves merging preprocessing steps and calculating the center of structure
NxN square metrix, N is no. of data points for cluster. NxN gives dissimilarity between two objects)


```python
import numpy as np

# Define the data set with three categorical variables
data = np.array([['A', 'B', 'C'],
				['B', 'C', 'A'],
				['C', 'A', 'B'],
				['A', 'C', 'B'],
				['A', 'A', 'B']])

# Choose the number of clusters k
k = 2

# Initialize the modes for each cluster
modes = [['A', 'B', 'C'],
		['C', 'B', 'A']]

# Initialize the cluster assignments for each data objects
clusters = np.zeros(data.shape[0], dtype=int)
clusters_prev = np.zeros(data.shape[0], dtype=int)

# Iteratively assign data objects to
# their clusters and update the cluster modes
for i in range(10):
	# Assign data objects to the closest cluster
	for j, object in enumerate(data):
		distances = np.array([sum(object != mode) for mode in modes])
		clusters[j] = np.argmin(distances)

	# Update the cluster modes
	for j in range(k):
		# Use np.unique to find the mode for categorical data
		values, counts = np.unique(data[clusters == j, :], axis=0, return_counts=True)
		modes[j] = values[np.argmax(counts)]

	# Check if the cluster assignments have converged
	if (clusters == clusters_prev).all():
		break

	# Store the current cluster assignments
	clusters_prev = clusters

# Print the cluster assignments for each data point
print("The cluster assignments for each data object: ", clusters)

# Print the modes for each cluster
print("Modes for each cluster: ", modes)
```

```python
#Importing necessary libraries
import pandas as pd
import numpy as np
from kmodes.kmodes import KModes
import matplotlib.pyplot as plt
%matplotlib inline

#Elbow curve to find the optimal value of K

cost = []

K = range(1,5)
for k in K:
  kmode = KModes(n_clusters=k, init='random', n_init=5, verbose=1)
  kmode.fit_predict(data)
  cost.append(kmode.cost_)


plt.plot(K, cost, 'x-')
plt.xlabel('Number of Clusters')
plt.ylabel('Cost')
plt.title('Elbow Method')
plt.show() #Elbow like shape at cluster 2 and 3 lets consider number 2
```
