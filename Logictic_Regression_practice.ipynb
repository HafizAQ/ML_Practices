```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.utils import shuffle

# Define the data for each column
# Replace these with your actual data
time_spent = [10, 20, 15, 25, 30]
reviews_read = [5, 10, 3, 7, 12]
product_rating = [4, 5, 3, 4, 5]
discount_percentage = [10, 20, 5, 15, 25]
prime_member = [1, 0, 1, 1, 0]  # 1 for True, 0 for False
purchased = [1, 1, 0, 1, 0]  # 1 for True, 0 for False

df = pd.DataFrame({
    "Time_Spent": time_spent,
    "Reviews_Read": reviews_read,
    "Product_Rating": product_rating,
    "Discount_Percentage": discount_percentage,
    "Prime_Member": prime_member,
    "Purchased": purchased
})

df = shuffle(df, random_state=42).reset_index(drop=True)
df.head()
```

```python
# Predictor Features (X) and Target(y)
X = df.drop("Purchased", axis=1)
y = df["Purchased"]

# Train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Logistic Regression Model
model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# Model Evaluation & print
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

accuracy, classification_rep
```

```python
import numpy as np

def predict_purchase_probability(time_spent, num_reviews, product_rating, discount, prime_status):
    # Given logistic regression coefficients
    beta_0 = -3.894
    beta_1 = 0.0435  # Time Spent
    beta_2 = 0.0201  # Number of Reviews Read
    beta_3 = 0.2802  # Product Rating
    beta_4 = 0.0450  # Discount Percentage
    beta_5 = 0.7629  # Prime Membership

    # Compute log-odds
    log_odds = (beta_0 + beta_1 * time_spent + beta_2 * num_reviews +
                beta_3 * product_rating + beta_4 * discount + beta_5 * prime_status)

    # Convert log-odds to probability using logistic function
    purchase_probability = 1 / (1 + np.exp(-log_odds))

    return purchase_probability

# Example customer data
customer_data = {
    "Time_Spent": 30,       # seconds on product page
    "Num_Reviews": 20,      # number of reviews read
    "Product_Rating": 4.5,  # product rating (1-5)
    "Discount": 25,         # discount percentage
    "Prime_Status": 1       # 1 = Prime Member, 0 = Non-Prime
}

# Calculate Probability
probability = predict_purchase_probability(
    customer_data["Time_Spent"],
    customer_data["Num_Reviews"],
    customer_data["Product_Rating"],
    customer_data["Discount"],
    customer_data["Prime_Status"]
)

probability
```
